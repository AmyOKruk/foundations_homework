```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

response = requests.get('https://www.goodreads.com/list/show/6934.Science_Fiction_Books_by_Female_Authors')
doc = BeautifulSoup(response.text)
```

```python
items = doc.find_all('tr', itemtype="http://schema.org/Book")
import re

rows = []
for item in items:
    row = {}
    row['ranking']=(item.find('td', class_="number").text)
    row['title']=(item.find('a', class_="bookTitle").text.strip())
    row['author']=(item.find('span', itemprop="author").text.strip())
    row['score']=(item.find('a', onclick="Lightbox.showBoxByID('score_explanation', 300); return false;").text)
    row['voters']=(item.find('a', id=re.compile(r'.*')).text)
    row['ratings']=(item.find('span', class_="greyText smallText uitext").text.strip())
    rows.append(row)
    
rows
    

    
```

```python
df = pd.DataFrame(rows)
df.head()
```

```python
df.to_csv(r"goodreads.csv")
```

```python
df['score'] = df['score'].str.extract(r'score:(.*)')
```

```python
df['score'] = df['score'].str.replace(',', '')
```

```python
df['number_of_ratings'] = df['ratings'].str.extract(r'—(.*)\bratings')
```

```python
df['ratings'] = df['ratings'].str.extract(r'(.*).\bavg')
```

```python
df['author']=df['author'].replace(to_replace=r'\(.*\)', value='', regex=True)

```

```python
df['series'] = df['title'].str.extract(r'[(](.*),?#')
```

```python
df['series'] = df['series'].str.replace(',','')
```

```python
df['number_in_series'] = df['title'].str.extract(r'#(\d\d?)')
```

```python
df['title']=df['title'].replace(to_replace=r'[(](.*)[)]', value='', regex=True)
```
