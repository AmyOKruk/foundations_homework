```python
rows = []

for page_num in range(1, 10):
    url = f"https://www.epicurious.com/search/cucumbers?page={page_num}"
    response = requests.get(url)
    doc = BeautifulSoup(response.text)
    
    regexp = re.compile(r'.*-content-card')
    articles = doc.find_all(class_=regexp)
    
    for article in articles:
        row = {}
        row['name']= article.find('h4', class_="hed").text
        row['tag']= article.find('strong', class_="tag").text
        row['summary']= article.find('p', class_="dek").text
        try:
            row['rating']= article.find('span', itemprop="ratingValue").text
        except:
            pass
        try:
            row['make again']= article.find('dd', class_="make-again-percentage").text
        except:
            pass
        row['link']= article.find('h4', class_="hed").a['href']
        rows.append(row)

df = pd.DataFrame(rows)
```

```python
def scrape_page(row):
    url = f"https://www.epicurious.com/" + row['link']
    print("Scraping", url)
    
    response = requests.get(url)
    doc = BeautifulSoup(response.text)
    
    page = {}
    steps = doc.find_all('li', class_="preparation-step")
    page['directions'] = '\n'.join([step.text for step in steps])
    ingredients = doc.find_all('li', class_="ingredient")
    page['ingredients'] = '\n'.join([ingredient.text for ingredient in ingredients])
    tags = doc.find_all('dt', itemprop="recipeCategory")
    page['tags'] = '\n'.join([tag.text for tag in tags])
    return pd.Series(page)
```

```python
scraped_df = df.apply(scrape_page, axis=1)
```
